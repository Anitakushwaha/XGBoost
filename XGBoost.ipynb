```python
data['Price']=boston.target
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
X,Y=data.iloc[:,:-1],data.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
data_dmatrix=xgb.DMatrix(data=X,label=Y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
xg_reg=xgb.XGBRegressor(objective='reg:linear',colsample_bytree=0.3,learning_rate=0.1,max_depth=5,alpha=10,n_estimators=10)
```

```python
xg_reg.fit(X_train,Y_train)
preds=xg_reg.predict(X_test)
```

```python
rmse=np.sqrt(mean_squared_error(Y_test,preds))
print("RMSE:%f"%rmse)
```

# The RMSE for the price prediction came out to be around 10.5 per $1000

# K fold Cross Validation using XGBoost

```python
params = {"objective":"reg:linear",'colsample_bytree': 0.3,'learning_rate': 0.1,
                'max_depth': 5, 'alpha': 10}

cv_results = xgb.cv(dtrain=data_dmatrix, params=params, nfold=3,
                    num_boost_round=50,early_stopping_rounds=10,metrics="rmse", as_pandas=True, seed=0)
```

```python
cv_results.head()
```

```python
cv_results["test-rmse-mean"].tail(1)
```

# RMSE of the price prediction has reduced as compared to last time and came out to be around 3.99 per $1000.

# Visualize Bossting Trees and Features Importance

```python
xg_reg=xgb.train(params=params,dtrain=data_dmatrix,num_boost_round=10)
```
